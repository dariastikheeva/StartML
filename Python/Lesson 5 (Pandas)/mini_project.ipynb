```python
# создадим датафрейм, который будет включать в себя: первые три строки 
# исходного датафрейма, строки 6-8 (включительно), строки 16-19 
# (включительно) и колонки: App, Size, Genres, Current Ver

df = playstore[['App', 'Size', 'Genres', 'Current Ver']]

df_1 = df.iloc[:3]
df_2 = df.iloc[5:8]
df_3 = df.iloc[15:19]

result = pd.concat([df_1, df_2, df_3])

import csv
result.to_csv('result.csv', sep=',', quoting=csv.QUOTE_ALL)
result2 = pd.read_csv('result.csv')
result2 
```

```python
# уберем дубликаты приложений из колонки App и сбросим индекс

unique_playstore = playstore.drop_duplicates(subset=['App']).reset_index(drop=True)
unique_playstore.head()
```

```python
# приведем колонки к стандартному виду

unique_playstore.columns = [column.lower().replace(' ', '_') for column in unique_playstore.columns]
del unique_playstore['unnamed:_0']
unique_playstore.head()
```

```python
# посчитаем долю бесплатных приложений (предварительно убрав дубликаты в колонке эпп)

unique_playstore['type'][unique_playstore['type'] == 'Free'].count() / len(unique_playstore['type'])
```

```python
# отберем из датафрейма только те приложения, которые относятся 
# к образовательной категории и с количеством отзывов пользователей более 1000

education_playstore = unique_playstore[(unique_playstore['category'] == 'EDUCATION') & (unique_playstore['reviews'] > 1000)]
education_playstore = education_playstore.reset_index(drop=True)
education_playstore = education_playstore.reset_index(drop=True)
education_playstore.head()
```

```python
# уберем лишние символы из колонки с ценой
# и переведем данные в тип float
unique_playstore['price'] = [float(i.replace('$', '')) for i in playstore['price']]
unique_playstore.head()
```

```python
playstore = pd.read_csv('playstore.csv')
playstore = playstore.drop_duplicates(subset=['App']).reset_index(drop=True)
playstore['Price'] = [float(i.replace('$', '')) for i in playstore['Price']]
playstore.columns = [column.lower().replace(' ', '_') for column in playstore.columns]
```

```python
# сделаем сводную таблицу по нашим данным
import numpy as np

table = pd.pivot_table(playstore,
               index=['category', 'type'],
               values=['price', 'rating', 'reviews'],
               aggfunc=[np.mean])
table
```

```python
# делаем таблицу красивее
table.columns = ['mean_price', 'mean_rating', 'mean_reviews']
table['mean_price'] = table['mean_price'].round(2)
table['mean_rating'] = table['mean_rating'].round(1)
table['mean_reviews'] = table['mean_reviews'].round(2)
table
```

```python
# сохраняем сводную таблицу в файл
table.to_csv('table.csv')
```
